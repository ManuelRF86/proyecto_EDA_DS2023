```python
"""Los primeros datos que estudiaremos, son los de faltas y tarjetas,
por tanto se crea una variable, solo con las columnas que nos interesa de momento.
Despues añadimos 2 columnas las cuales determinan cuantas faltas se necesitan en cada partido para que saquen una amarilla.
Las 2 columnas son para el equipo local y el visitante"""

total_amonestaciones = total_temporadas[['HomeTeam','AwayTeam','HF','AF','HY','AY','HR','AR','Temporada']]
total_amonestaciones['promedio_F_H/Y'] = round(total_amonestaciones['HF'] / total_amonestaciones['HY'],1)
total_amonestaciones['promedio_F_A/Y'] = round(total_amonestaciones['AF'] / total_amonestaciones['AY'],1)
total_amonestaciones
```

```python
total_amonestaciones['promedio_F_H/Y'] = total_amonestaciones['promedio_F_H/Y'].replace(np.inf, 0.8)  #sustiuimos el valor por 0.8 para que destaque en la gráfica y obtener así un dato superior
total_amonestaciones['promedio_F_A/Y'] = total_amonestaciones['promedio_F_A/Y'].replace(np.inf, 0.8)  #si los sustituyo por 0, le media bajará considerablente ya que le estamos diciendo que necesito 0 faltas para 1 tarjeta
```

```python
total_amonestaciones.info()
```

```python
"""Mediante un grouby, se crea 2 tablas(local y visitante) con los datos de cada equipo y cada temporada.
Los valores son la suma de faltas, amarillas y rojas"""

tabla_por_equipos_A = total_amonestaciones.groupby(['AwayTeam','Temporada'],as_index=False)[['AF','AY','AR',]].agg('sum')
tabla_por_equipos_A = tabla_por_equipos_A.rename(columns={'AwayTeam':'Equipo'})
tabla_por_equipos_A
```

```python
tabla_por_equipos_H = total_amonestaciones.groupby(['HomeTeam','Temporada'],as_index=False)[['HF','HY','HR']].agg('sum')
tabla_por_equipos_H = tabla_por_equipos_H.rename(columns={'HomeTeam':'Equipo'})
tabla_por_equipos_H
```

```python
"""Fusionamos las tablas anteriores,
Se crea varias columnas:
-Totales
-Promedio faltas por tarjeta
-Numero de temporadas de cada equipo en primera
Por último, se crea una variable con quedando solo los equipos con mas de 3 temporadas para que no se falseen alguns datos
"""

tabla_equipos = tabla_por_equipos_H.merge(tabla_por_equipos_A, on=('Equipo','Temporada'))
tabla_equipos['Total_F'] = tabla_equipos['HF'] + tabla_equipos['AF']
tabla_equipos['Total_Y'] = tabla_equipos['HY'] + tabla_equipos['AY']
tabla_equipos['Total_R'] = tabla_equipos['HR'] + tabla_equipos['AR']
tabla_equipos['Promedio_F/Y'] = round(tabla_equipos['Total_F'] / tabla_equipos['Total_Y'],1)


tabla_equipos['num_temp'] = tabla_equipos.groupby('Equipo')['Equipo'].transform('count') 

tabla_equipos_3temp = tabla_equipos[tabla_equipos['num_temp'] > 3]   
tabla_equipos_3temp
```

```python
"""En esta variable se obtiene una tabla con la suma de los datos de los equipos (separados por local o visitante)
en todas sus temporadas en primera y añadiendo el numero de temporadas.
Ademas se crea el promedio de faltas por tarjeta y la media de rojas por temporada.
En este caso, se quedan solo los equipos con mas de 5 temporadas en primera"""

tabla_equipos_HA = totales_por_temp = tabla_equipos.groupby(['Equipo','num_temp'],as_index=False)[['HF','HY','HR','AF','AY','AR']].aggregate('sum')
tabla_equipos_HA['YF_H'] = round(tabla_equipos_HA['HF'] / tabla_equipos_HA['HY'],1)
tabla_equipos_HA['YF_A'] = round(tabla_equipos_HA['AF'] / tabla_equipos_HA['AY'],1)
tabla_equipos_HA['MediaR_A'] = round(tabla_equipos_HA['AR'] / tabla_equipos_HA['num_temp'],1)
tabla_equipos_HA['MediaR_H'] = round(tabla_equipos_HA['HR'] / tabla_equipos_HA['num_temp'],1)
tabla_equipos_HA = tabla_equipos_HA[tabla_equipos_HA['num_temp'] > 5]
tabla_equipos_HA
```

```python
total_temp = pd.read_csv('data_limpio/total_temp.csv')
```

```python
"""Para sacar los ataque totales, ya que ese datos ha sido imposible de conseguir,
nos basamos en disparos realizados y recibidos, 
ya que al menos, para disparar hay que llegar cerca del área
y tras visualizar varios gráficos,
evaluamos que con tantos datos, es un buen datos para medir los menaltis, según los acercamientos al área."""

total_ataques = total_temp[['HomeTeam','AwayTeam','HS','AS','Temporada']]
total_ataques
```

```python
"""Se crean 2 tablas(local y visitante),
con la suma de todos los disparos realizados y recibidos de cada equipo en cada temporada.
Renombramos la columna en este caso como disparos recibidos.
La forma de hacer el groupby, al igual que en una tabla anterior pero en ese caso de faltas, de la siguiente manera:
Por cada equipo local, los HS(home shoot) serán los disparos realizados,
los AS(away shoot) sern los disparos recibidos.
Para el visitante, los hacemos a la inversa. 
"""

ataques_H = total_temp.groupby(['HomeTeam','Temporada'],as_index=False)[['HS','AS']].agg('sum')   #añado la columna contraria al local para obtener los tiros recibidos
ataques_H = ataques_H.rename(columns={'HomeTeam':'Equipo','AS':'RS_H'})
ataques_H
```

```python
ataques_A = total_temp.groupby(['AwayTeam','Temporada'],as_index=False)[['AS','HS']].agg('sum')     #añado la columna contraria al visitante para obtener los tiros recibidos
ataques_A = ataques_A.rename(columns={'AwayTeam':'Equipo','HS':'RS_V'})
ataques_A
```

```python
"""Se fusionan las 2 tablas y creamos 2 columnas con las sumas de realizados y recibidos"""
ataques_totales = ataques_A.merge(ataques_H, on=('Equipo','Temporada'))
ataques_totales['realizados'] = ataques_totales['AS'] + ataques_totales['HS']
ataques_totales['recibidos'] = ataques_totales['RS_V'] + ataques_totales['RS_H']
ataques_totales
```

```python
ataques_totales.to_csv('data_limpio/total_ataques.csv')
```

```python
"""Para empezar a trabajar con los penaltis,
leemos la tabla creada y exportada mediante webscrapping
"""
penaltis_FC = pd.read_csv('data_limpio/penaltis_FC.csv')
```

```python
"""Nos encontramos con el problema de que los nombres de los equipos,
están representados diferentes en ambas tablas, pr lo que necesitamos modificar los de una de ellas.
Están dividido en 7 lineas para tener mejo visualización y no cometer errores"""

penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['RCD Mallorca','RC Deportivo de Corunha','Athletic Bilbao','FC Villarreal','RCD Espanyol','Real Sociedad'],['Mallorca','La Coruna','Ath Bilbao','Villarreal','Espanol','Sociedad'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['FC Barcelona','Cádiz CF','Valência CF','Getafe CF','Celta de Vigo','Sevilla FC'],['Barcelona','Cadiz','Valencia','Getafe','Celta','Sevilla'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['Real Bétis','CA Osasuna','Real Madrid CF','Atlético de Madrid','Real Zaragoza','SD Eibar'],['Betis','Osasuna','Real Madrid','Ath Madrid','Zaragoza','Eibar'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['Málaga CF','Racing Santander','Deportivo Alavés','Levante UD','Gimnàstic de Tarragona'],['Malaga','Santander','Alaves','Levante','Gimnastic'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['Recreativo de Huelva','Real Valladolid CF','Real Murcia CF','UD Almería','CD Numância','Elche CF'],['Recreativo','Valladolid','Murcia','Almeria','Numancia','Elche'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['Real Sporting de Gijón','CD Tenerife','Xerez CD','Hércules CF','Granada CF','Rayo Vallecano'],['Sp Gijon','Tenerife','Xerez','Hercules','Granada','Vallecano'])
penaltis_FC['Equipo'] = penaltis_FC['Equipo'].replace(['Córdoba CF','UD Las Palmas','CD Leganés','Girona FC','SD Huesca'],['Cordoba','Las Palmas','Leganes','Girona','Huesca'])
```

```python
"""Ahora sí, se pueden fusionar ambas tablas por las columnas coincidentes"""

ataque_y_penaltis = ataques_totales.merge(penaltis_FC, on=('Equipo','Temporada'))
ataque_y_penaltis = ataque_y_penaltis.drop('Unnamed: 0',axis=1)
ataque_y_penaltis = ataque_y_penaltis.rename(columns={'penaltis': 'penaltis_favor'})
```

```python
"""volvemos a hacer lo mismo para sacar el promedio, 
Contamos cuando son 0 penalti, como 0.8 para obtener una mayor "penalizacion":
Como entre 0 no podemos dividir, ponemos 0.8 para que en graficos, destaque ligeramente por encima del 1
- 1 penalti y 100 ataques: necesitaría 100 ataques para que piten un penlati
- 0 penalti (0.8) y 100 ataques: necesitaría 125 ataques para que le pitaran un penalti.
Quizas no sea la forma mas real pero es importante que el valor de mas peso, se encuentre en el promedio.
A la hora de representar la cantidad de penaltis, los modifico para que aparezca 0 si procede"""

ataque_y_penaltis['penaltis_contra'] = ataque_y_penaltis['penaltis_contra'].replace(0,0.8)
ataque_y_penaltis['penaltis_favor'] = ataque_y_penaltis['penaltis_favor'].replace(0,0.8)
```

```python
"""Creamos el promedio de ataque/penaltis tanto realizados como recibidos,
de cada equipo por temporada."""

ataque_y_penaltis['promedio_PC'] = round((ataque_y_penaltis['RS_H'] + ataque_y_penaltis['RS_V']) / ataque_y_penaltis['penaltis_contra'],1)
ataque_y_penaltis['promedio_PF'] = round((ataque_y_penaltis['AS'] + ataque_y_penaltis['HS']) / ataque_y_penaltis['penaltis_favor'],1)
```

```python
ataque_y_penaltis['num_temp'] = ataque_y_penaltis.groupby('Equipo')['Equipo'].transform('count')
ataque_y_penaltis
```

```python
ataque_y_penaltis['penaltis_contra'] = ataque_y_penaltis['penaltis_contra'].replace(0.8,0)
ataque_y_penaltis['penaltis_favor'] = ataque_y_penaltis['penaltis_favor'].replace(0.8,0)
```

```python
ataque_y_penaltis.to_csv('data_limpio/ataques_y_penaltis.csv')
```

```python
"""Creamos una tabla con la sumas y promedios de penaltis a favor y en contra,
de la suma de todas las temporadas que cada equipos tiene en primera."""

promedio_total = ataque_y_penaltis.groupby(['Equipo','num_temp'],as_index=False)[['RS_V','RS_H','AS','HS','penaltis_contra','penaltis_favor']].aggregate('sum').round()
promedio_total['promedio_PC'] = round((promedio_total['RS_H'] + promedio_total['RS_V']) / promedio_total['penaltis_contra'],1)
promedio_total['promedio_PF'] = round((promedio_total['AS'] + promedio_total['HS']) / promedio_total['penaltis_favor'],1)
promedio_total
```
